```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
from sklearn.metrics import accuracy_score
%matplotlib inline

iris = load_iris()
X = iris.data
y = iris.target

# Split data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
## Random Forest
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)

print(accuracy_score(y_test, y_pred))

species_idx = rfc.predict([[3, 5, 4, 2]])[0]
iris.target_names[species_idx]

## visualization
sns.barplot(x=rfc.feature_importances_, y=iris.feature_names)

plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```

```python
## Bagging
bag = BaggingClassifier(n_estimators=100)
bag.fit(X_train, y_train)

y_pred = bag.predict(X_test)

print(accuracy_score(y_test, y_pred))


## AdaBoost
ada = AdaBoostClassifier(n_estimators=100)
ada.fit(X_train, y_train)

y_pred = ada.predict(X_test)

print(accuracy_score(y_test, y_pred))

## Visualization
sns.barplot(x=ada.feature_importances_, y=iris.feature_names)

plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()


## Gradient Boosting
gbc = GradientBoostingClassifier(n_estimators=100)
gbc.fit(X_train, y_train)

y_pred = gbc.predict(X_test)

print(accuracy_score(y_test, y_pred))

## Visualization
sns.barplot(x=gbc.feature_importances_, y=iris.feature_names)

plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()

```
